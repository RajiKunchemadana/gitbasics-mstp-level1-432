### Line plot

```python
import matplotlib.pyplot as p
x=[2,10,6,8]
y=[1,8,5,9]
p.plot(x,y)
p.show()
```

```python
# by using range display lineplot
#import matplotlib.pyplot as p
x=list(range(1,20,2)) # [expression loop condtion]
y=list(range(5,25,2))
p.plot(x,y,"go")# circle=o,square=s
p.show()
```

```python
# line plot by using list comphresion
x=[i for i in range(1,11)]
y=[j**0.5 for j in x]
p.plot(x,y,"go")
p.grid()
```

##### Label names and title for a graph

* lib.xlabel('x-axis')
* lib.ylabel('y-axis')
* lib.title("sqrat of numbers")

```python
x=[2,4]
y=[10,5]
p.plot(x,y)
p.xlabel('x-axis')
p.ylabel('y-axis')
p.title("line plot")
p.grid()
p.show()

```

# Bar graph

```python
import matplotlib.pyplot as p
x=list(range(1,10,3))
y=list(range(11,20,3))
p.bar(x,y)
```

```python
# scatter plot
import matplotlib.pyplot as p
x=list(range(1,50,2))
y=list(range(11,60,2))
p.scatter(x,y)
p.title("range")
```

```python
# histogram plot
import matplotlib.pyplot as p
x=[1,3,0,3]
p.hist(x)
```

```python
# draw a plot for time and distance by using list comhresion
time=[i for i in range(1,20,2)]
distance=[2*j for j in time]
#p.plot(time,distance)
p.bar(time,distance)
#p.scatter(time,distance)
#p.hist(time)
p.xlabel('time')
p.ylabel('distance')
p.title("line plot")
p.grid()
p.show()
```

## Getting images by using matplotlib

```python
import matplotlib.pyplot as pl
import matplotlib.image as m
image=m.imread('ram.jpg')
pl.imshow(image)
```

```python
### Applying psudo colors to an image
import matplotlib.pyplot as pl
import matplotlib.image as m
x=m.imread('ram.jpg')
c=x[:,:,0]## Apply psudo color
pl.imshow(c)
```

```python
import matplotlib.pyplot as pl
import matplotlib.image as m
x=m.imread('ram.jpg')
c=x[:,:,0]## Apply psudo color
pl.imshow(c)
pl.imshow(c,cmap='hot')
```
