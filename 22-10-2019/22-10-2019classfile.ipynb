## matplotlib ticks

* ticks is values within a coordinates

```python
import matplotlib.pyplot as pl
x=[5,10,4,7,3,9,20]
y=[3,5,2,7,6,8,20]
pl.plot(y,x)
pl.title("ticks")
pl.xlabel("natural numbers")
pl.ylabel("values ")
pl.show()
```

```python
import matplotlib.pyplot as pl
import numpy as np
pl.xticks(np.arange(1,11))
pl.yticks(np.arange(5,50,5))
pl.plot(x,y)
pl.title("arange")
pl.xlabel("natural numbers")
pl.ylabel("values")
pl.grid()
pl.show()
```

# representing a sine wave
* a=sine(wt)
* a=amplitude
* w=frequency
* t=time

```python
import numpy as np
import matplotlib.pyplot as pl
time=np.arange(1,10,.1)
amplitude=np.sin(time)
pl.plot(time,amplitude)
pl.title("sine wave")
pl.xlabel("time")
pl.ylabel("amplitude")
pl.grid()
pl.show()
```
