# seaborn

```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
help(sns.factorplot)
```

```python
sns.factorplot(x='total_bill',y='day',data=tips)
```

```python
sns.factorplot(x='total_bill',y='day',data=tips,hue='sex')
```

```python
sns.factorplot(x='total_bill',y='day',data=tips,hue='sex',markers=['x','s'])# markers is based on the or depends on hue
```

```python
sns.factorplot(x='total_bill',y='day',data=tips,hue='sex',markers=['x','s'],dodge=True)
```

```python
from numpy import median
sns.factorplot(x='total_bill',y='day',data=tips,hue='sex',markers=['x','s'],dodge=True,estimator=median)
```

```python
sns.factorplot(x='sex',y='tip',data=tips,hue='day',markers=['x','s','o','d'])
```

```python
help(sns.pointplot)
```

```python
sns.pointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.pointplot(x='total_bill',y='day',data=tips,hue='smoker')
```

```python
sns.pointplot(x='total_bill',y='day',data=tips,hue='smoker',markers=['x','o'])
```

```python
sns.pointplot(x='total_bill',y='sex',data=tips,hue='day',markers=['x','o','s','d'],order=['Male','Female'])
```

```python
sns.pointplot('total_bill','sex',data=tips,hue='day',color='g')
```

```python
sns.pointplot('total_bill','sex',data=tips,hue='day',palette='PuBuGn')
```

```python
sns.pointplot('total_bill','sex',data=tips,hue='day',markers=['x','s','o','d'],linestyles=[':','-.','--','-'])
```

```python
sns.pointplot('tip','day',data=tips,hue='sex')
```

```python
sns.pointplot('tip','day',data=tips,hue='sex',join=False)
```

```python
sns.pointplot('tip','day',data=tips,hue='sex',join=True)
```

```python
sns.pointplot('tip','day',data=tips,hue='sex',join=False,dodge=True,estimator=median)
```

```python
import seaborn as sns
sns.color_palette()
```

```python
sns.palplot(sns.color_palette())
```

```python
c=sns.color_palette('RdBu')
sns.palplot(c)
```

```python
c=sns.color_palette('RdBu',n_colors=20)
sns.palplot(c)
```

```python
c=sns.color_palette('RdBu',10)
sns.palplot(c)
```

```python
sns.palplot(sns.color_palette('ch:4'))
```

```python
sns.palplot(sns.color_palette('ch:4,r=1,l=0'))
```

```python
sns.set()
cc=sns.cubehelix_palette()
cc
```

```python
sns.palplot(cc)
```

```python
sns.palplot(sns.cubehelix_palette(reverse=True))
```

```python
sns.palplot(sns.cubehelix_palette(start=2,rot=-1))
```

```python
sns.palplot(sns.cubehelix_palette(start=2))
```

```python
sns.palplot(sns.cubehelix_palette(start=2,rot=-1,reverse=True))
```

```python
import matplotlib.pyplot as plt
sns.heatmap(sns.cubehelix_palette())
```

```python
import numpy as np
x=np.arange(16).reshape(4,4)
cc=sns.cubehelix_palette()
sns.heatmap(x,cmap=cc)
```

```python
x=np.arange(16).reshape(4,4)
cc=sns.cubehelix_palette(dark=-2)
sns.heatmap(x,cmap=cc)
```

```python
x=np.arange(16).reshape(4,4)
cc=sns.cubehelix_palette(light=2)
sns.heatmap(x,cmap=cc)
```

```python
x=np.arange(16).reshape(4,4)
cc=sns.cubehelix_palette(light=2,as_cmap=True)
sns.heatmap(x,cmap=cc)
```
