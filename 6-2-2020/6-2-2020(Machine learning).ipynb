```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ad_df=pd.read_csv("advertisement.csv")

ad_df.head()
```

```python
ad_df
```

```python
ad_df.tail()
```

```python
ad_df.drop("Unnamed: 0",axis=1,inplace=True)
ad_df.head()
```

```python
x=ad_df.iloc[:,0:3].values
#print(x)
y=ad_df.iloc[:,3:].values
#print(y)
print(x.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
```

```python
lin.fit(x,y)
```

```python
y_predict=lin.predict(x)
y_predict
```

```python
lin.predict([[180.8,10.8,58.4]])
```

```python
lin.predict([[18,44.4,58.8]])
```

```python
plt.scatter(ad_df['radio'],y,c='r',label='before prediction')
plt.plot(x,y_predict,c='y',label='after prediction with linear regression')
plt.xlabel('cost of tv,radio newspaper')
plt.ylabel('sales')
plt.title('linear regression')
plt.legend(loc='best')
plt.show()
```
